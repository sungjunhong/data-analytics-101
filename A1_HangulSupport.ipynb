### 구글 코랩에서 한글 사용하기 (feat. 나눔폰트)

```python
import matplotlib.pyplot as plt
import numpy as np

years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(years, values)
plt.xlabel('연도')
plt.ylabel('판매량')
plt.show()
```
