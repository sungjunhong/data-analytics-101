### 산점도 만들기

```python
from google.colab import drive

drive.mount('/content/gdrive')
path = '/content/gdrive/MyDrive/data-analytics-101/data/'
```

```python
import pandas as pd
mpg = pd.read_csv(path + 'mpg.csv')
```

```python
import plotly.express as px

px.scatter(data_frame = mpg, x = 'cty', y = 'hwy', color = 'drv')
```

### 막대 그래프 만들기

```python
# 자동차 종류별 빈도 구하기
df = mpg.groupby('category', as_index = False) \
        .agg(n = ('category', 'count'))
df
```

```python
px.bar(data_frame = df, x = 'category', y = 'n', color = 'category')
```

### 선 그래프 만들기

```python
# economics 불러오기
economics = pd.read_csv(path + 'economics.csv')                  
```

```python
px.line(data_frame = economics, x = 'date', y = 'psavert')  # psavert : personal saving rate
```

### 상자 그림 만들기

```python
px.box(data_frame = mpg, x = 'drv', y = 'hwy', color = 'drv')
```

## 인터렉티브 그래프 저장하기

### HTML 파일로 저장하기

```python
fig = px.scatter(data_frame = mpg, x = 'cty', y = 'hwy')

fig.write_html('/content/gdrive/MyDrive/data-analytics-101/scatter_plot.html')
```
