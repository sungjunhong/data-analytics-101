##**'함수' 이해하기**

image.png

- 함수 호출(function call): 함수를 사용한다는 것을 함수를 호출한다고 표현함
- 매개변수(parameter): 함수를 호출할 때 함수의 입력으로 넣은 자료
- 리턴값(return value): 함수를 호출해서 최종적으로 나오는 결과값

- 데이터 분석은 **함수를 이용해서 변수를 조작하는 일**이라고 할 수 있음.
- 파이썬의 다양한 표준 라이브러리와 외부 패키지(pandas, numpy, seaborn, etc)에 포함된 함수의 기능과 구현 방법을 익혀 데이터를 의도한대로 분석


### 함수 호출하기

````python
# 변수 만들기
x = [1, 2, 3]
x
````

````python
sum(x)
````

````python
max(x)
````

````python
min(x)
````

### 함수의 결과물로 새 변수 만들기

````python
x_sum = sum(x)
x_sum
````

````python
x_max = max(x)
x_max
````

### 함수 만들기

````python
def fn_sum(x):
    ret = 0;
    for i in x:
        ret = ret + i
    return ret
````

````python
fn_sum(x)
````

###**`seaborn`의 `titanic` 데이터로 그래프 그리기**

#### `seaborn`의 `load_dataset()`로 `titanic` 데이터 불러오기

````python
df = sns.load_dataset('titanic')
df
````

#### 함수의 다양한 기능 이용하기

````python
sns.countplot(data = df, x = 'sex')
````

````python
sns.countplot(data = df, x = 'class')                 # x축 class
````

````python
sns.countplot(data = df, x = 'class', hue = 'alive')  # x축 class, alive별 색 표현
````

````python
sns.countplot(data = df, y = 'class', hue = 'alive')  # y축 class, alive별 색 표현
````

[꿀팁] 함수 사용법이 궁금할 땐 도움말 명령(?)을 이용하자.

````python
sns.countplot?
````

### 모듈

- `모듈(module)`: 여러 변수와 함수를 가지고 있는 집합체로 표준 모듈과 외부 모듈로 구분

image.png

####**`패키지명.모듈명.함수명()`으로 함수 사용하기**

```
import sklearn.metrics
sklearn.metrics.accuracy_score()
```




####**`모듈명.함수명()`으로 함수 사용하기**

```
from sklearn import metrics
metrics.accuracy_score()
```

####**`함수명()`으로 함수 사용하기**

```
from sklearn.metrics import accuracy_score
accuracy_score()
```

[꿀팁] as로 패키지명, 모듈명, 함수명 약어 지정하기

```
import sklearn.metrics as met 
met.accuracy_score()
```
```
from sklearn import metrics as met 
met.accuracy_score()
```
```
from sklearn.metrics import accuracy_score as accuracy 
accuracy()
```
