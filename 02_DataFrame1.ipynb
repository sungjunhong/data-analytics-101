# **둘째마당. 데이터 프레임의 세계**

## 데이터 프레임 이해하기

**데이터 프레임(data frame)**: 행과 열로 구성된 표 모양 pandas 패키지의 자료구조

연봉에 영향을 미치는 요인을 파악하기 위한 원자료

image.png

**열(column):** 변수 또는 컬럼이라고 하며, 속성이나 특징을 구분하기 위해 열 제목(변수명, 속성명)을 사용

image.png

**행(row):** 샘플(sample), 케이스(case)라고도 하며 행의 개수가 데이터의 개수를 의미함. 데이터의 순서를 구분하기 위해 행 인덱스를 사용.

image.png

image.png

* 데이터가 크다는 것는 행이 많거나 열이 많은 경우를 말하는데, 그 성격에 따라 접근해야하는 방식이 다름.

image.png

**1. `pandas` 패키지 로드하기**



```python
import pandas as pd
```

**2. 데이터 프레임 만들기**





```python
df = pd.DataFrame({'name'    : ['김지훈', '이유진', '박동현', '김민지'],
                   'english' : [90, 80, 60, 70],
                   'math'    : [50, 60, 100, 20]})
df
```

### 데이터 프레임으로 분석하기


#### 특정 변수의 값 추출하기



데이터 프레임 이름 뒤에 대괄호(`[ ]`)입력하고 문자 형태의 열 제목(변수명, 특징명)을 입력하여 참조

```python
df['english']
```

#### 변수의 값으로 합계 구하기



```python
sum(df['english'])
```

```python
sum(df['math'])
```

#### 변수의 값으로 평균 구하기

```python
sum(df['english']) / 4  # 영어 점수 평균
```

```python
sum(df['math']) / 4     # 수학 점수 평균
```
